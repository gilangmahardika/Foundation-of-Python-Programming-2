# Use Case: High Rating Movies Criteria

Hints:
* Let’s set imdb_score benchmark as High-Rating movie with imdb_score > 7.5
* Create new variables “score_group” (“High” if imdb_score > 7.5 and “Low” if imdb <= 7.5)
* Analyze at least 5 variables from which you can start figure out criteria of high rated movie
* Analyse 5 more variables to strengthen criteria of high rated movies

```python
new_cat_score = []

for row in imdb['imdb_score']:
    if row > 7.5:
        new_cat_score.append('High')
    else:
        new_cat_score.append('Low')
```

```python
imdb['score_group'] = new_cat_score
```

```python
imdb.head()
```

```python
imdb[imdb['score_group'] == 'High'].describe(include = 'all').transpose()
```

```python
imdb[imdb['score_group'] == 'Low'].describe(include = 'all').transpose()
```
